```python
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import time
```

```python
# driver = webdriver.Chrome('/Users/erthh/Documents/Web scraping/chromedriver')
driver.get("https://reverb.com/marketplace/electric-guitars?make=fender&page=100")
time.sleep(5)
html_source = driver.page_source
```

# Web scraping and data gathering function

```python
def get_info_dataframe(URL,brand):
    driver.get(URL)
    time.sleep(5)
    html_source = driver.page_source 
    soup = BeautifulSoup(html_source,'lxml')
    #get product_name
    all_product_name = soup.find_all('h4', class_='grid-card__title')
    product_name_list = [i.text for i in all_product_name]
    
    #get product price 
    all_product_price = soup.find_all('span', class_='price-display')
    product_price_list = [float(i.text.replace('$','').replace(',','')) for i in all_product_price]

    #get product condition 
    all_product_condition = soup.find_all('div', class_='condition-indicator__label')
    product_condition_list = [i.text for i in all_product_condition]
    
    #get producr URL 
    all_product_url = soup.find_all('a', class_='grid-card__inner')
    product_URL_list = [i.get('href') for i in all_product_url]
    
    #combine everythine into single dataframe
    data_dict = {'Product Name':product_name_list,'Price':product_price_list,'Condition':product_condition_list,'URL':product_URL_list,'Brand':Brand}
    df = pd.DataFrame(data_dict)
    return df
```

# Create a new data frame from specific guitar brand and how many page we want

```python
make_list = ['fender','ibanez','paul-reed-smith','gibson','suhr']
page_amount = 10
main_df = []
for i in make_list:
    for j in range(page_amount):
        URL = f'https://reverb.com/marketplace/electric-guitars?make={i}&page={j+1}'
        Brand = i 
        dataframe_temp = get_info_dataframe(URL,Brand)
        main_df.append(dataframe_temp)

new_df = pd.concat(main_df)
```

# Export scraped data to CSV file for later analysis

```python
#export to CSV file 
new_df.to_csv("guitars_market_analysis.csv", encoding='utf-8')
```
